```python
import matplotlib
import matplotlib.ticker as ticker
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
matplotlib.rcParams['pdf.fonttype'] = 42

df=pd.read_csv("Accuracy_1.csv")
for i in range(2,300):
       filename="Accuracy_"+str(i)+".csv"
       df+=pd.read_csv(filename)
df=df/299
df.describe()
# df.index = df.index + 1
# df.index = np.arange(1, len(df)+1)
# plot1 = df.plot('Epochs' , fontsize=12)
plot1 = df[['Train','Vallidation']].plot( fontsize=16
                                         , linewidth = 5)

plot1.set_ylabel('Accuracy %', fontsize=16) 
plot1.set_xlabel('Number of Epochs',fontsize=16)
plot1.legend(loc='upper left', fontsize=16)
plot1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))



ax1 = plot1.twinx()



# df = pd.DataFrame(a)

df['Energy consumption'] = a

# df = df.iloc[2:]
ax2= df[['Energy consumption']].plot(width=0.8, ax=ax1 ,kind='bar',secondary_y=True,legend=False, mark_right=False, color='g',stacked=False, alpha=0.4 , fontsize=16 )
ax2.set_ylabel('Energy (mJ)' , fontsize=16)

ax2.set_xticklabels(range(1,21))
ax2.legend(loc='lower center', fontsize=16)
# ax2.figure.subplots_adjust(bottom=0.4)
# ax2.axes.set_ylim(0,1000)
# for label in ax2.xaxis.get_ticklabels()[:2]:
#     label.set_visible(False)
# ax2.xaxis.set_minor_locator(ticker.AutoMinorLocator(1))
# ax2.xaxis.set_major_locator(ticker.MaxNLocator(nbins=4))
# print(ax2.xaxis.get_xticklabels()[1::2])
plt.setp(plot1.get_xticklabels()[::2], visible=False)


# plt.setp(ax2.get_xticklabels()[1::2], visible=False)
         
    
# ax2.figure.tight_layout()
ax2.figure.savefig("Accuracy-HAR.pdf",bbox_inches='tight')
```
